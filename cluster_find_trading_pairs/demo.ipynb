```python
from selection_criteria import PairsSelection
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("../data/raw_data.csv", index_col=0)
raw_data.head()
```

```python
pair_selection = PairsSelection(leg1_name="DGX", leg1=raw_data.loc[:, "DGX"], leg2_name="MTD", leg2=raw_data.loc[:, "MTD"])
print(pair_selection.is_eligible())
print(pair_selection.adf_pvalue)
print(pair_selection.hedge_ratio)
print(pair_selection.hurst_exponent)
print(pair_selection.half_life)
print(pair_selection.zero_cross)
```

```python
trading_pairs = pd.DataFrame(columns=["leg1", "leg2", "coint_t", "coint_pvalue", "hedge_ratio", "adf_pvalue",
                                      "hurst_exponent", "half_life", "zero_cross"])

columns = raw_data.columns[:50]
print("Eligible trading pairs:")
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        s1 = columns[i]
        s2 = columns[j]
        pair_selection = PairsSelection(leg1_name=s1, leg1=raw_data.loc[:, s1], leg2_name=s2, leg2=raw_data.loc[:, s2])
        if(pair_selection.is_eligible()):
            print(f"{s1} {s2}")
            trading_pair = [pair_selection.leg1_name, pair_selection.leg2_name,
                            pair_selection.coint_t, pair_selection.coint_pvalue,
                            pair_selection.hedge_ratio, pair_selection.adf_pvalue,
                            pair_selection.hurst_exponent, pair_selection.half_life, pair_selection.zero_cross]

            trading_pairs.loc[len(trading_pairs.index)] = trading_pair

# trading_pairs.to_csv("./trading_pairs.csv")
```

```python
trading_pairs.to_csv("../trading_pairs.csv")
```
